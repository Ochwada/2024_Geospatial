```python
import os
import subprocess

def convert_notebook_to_pdf(notebook_path, output_dir=None):
    """
    Converts a Jupyter Notebook to PDF.
    
    Parameters:
    - notebook_path: Path to the Jupyter Notebook (.ipynb).
    - output_dir: Directory where the PDF will be saved. Defaults to the notebook's directory.
    
    Returns:
    - Path to the generated PDF.
    """
    if not notebook_path.endswith(".ipynb"):
        raise ValueError("The file must be a Jupyter Notebook (.ipynb).")
    
    # Resolve absolute paths
    notebook_path = os.path.abspath(notebook_path)
    if not os.path.exists(notebook_path):
        raise FileNotFoundError(f"Notebook file not found: {notebook_path}")

    notebook_dir = os.path.dirname(notebook_path)
    output_dir = output_dir or notebook_dir

    # Ensure output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    pdf_path = os.path.join(output_dir, os.path.splitext(os.path.basename(notebook_path))[0] + ".pdf")

    try:
        # Convert notebook to PDF
        subprocess.run(
            [
                "jupyter", "nbconvert",
                "--to", "pdf",
                "--output-dir", output_dir,
                notebook_path
            ],
            check=True
        )
        print(f"PDF successfully generated: {pdf_path}")
    except subprocess.CalledProcessError as e:
        print("Error occurred while converting notebook to PDF.")
        print(f"Error: {e}")
        print("Ensure Pandoc and LaTeX are installed correctly.")
    except Exception as e:
        print("An unexpected error occurred.")
        print(f"Error: {e}")
    
    return pdf_path

# Example usage
if __name__ == "__main__":
    notebook_path = "01_geopandas_read.ipynb"  # Replace with your notebook path
    output_dir = "output"  # Optional, replace with your desired output directory
    try:
        convert_notebook_to_pdf(notebook_path, output_dir)
    except Exception as e:
        print(f"Error: {e}")
```
